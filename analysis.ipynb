```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

# 1. mumi mlumat
print(df.head())
print(df.info())
print(df.describe())

# 2. Fraud drcsi
fraud_rate = df["is_fraud"].mean() * 100
print("Fraud Rate:", fraud_rate)

# 3. Kart tipin gr xrclr
df.groupby("card_type")["transaction_amount"].mean().plot(kind="bar")
plt.title("Avg Transaction Amount by Card Type")
plt.savefig("images/card_type_amount.png")
plt.show()

# 4. Saat zr mliyyatlar
df.groupby("transaction_hour")["transaction_amount"].sum().plot(kind="line")
plt.title("Total Transactions by Hour")
plt.savefig("images/hour_amount.png")
plt.show()

# 5. Merchant Category zr risk
df.groupby("merchant_category")["is_fraud"].mean().plot(kind="bar")
plt.title("Fraud Rate by Merchant Category")
plt.savefig("images/fraud_by_category.png")
plt.show()

# 6. hrlr gr xrclr
df.groupby("city")["transaction_amount"].sum().plot(kind="pie", autopct='%1.1f%%')
plt.title("Transaction Distribution by City")
plt.savefig("images/city_pie.png")
plt.show()
```
